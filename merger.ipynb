```python
final_labels = {}

for x in xrange(130,141):
    with open("/home/lroot/code/lshtc/merge/res-"+str(x)+"-1.csv") as f:
        lines = f.readlines()
        i = 0
        for line in lines:
            
            line = line[1:len(line)-2]
            labels = line.split(',')
            if final_labels.has_key(i):
                final_labels[i].extend(labels)
            else:
                final_labels[i] = labels
            i+=1
```
